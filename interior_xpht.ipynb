```python
def compute_xpht_interior(exterior_xpht):
    '''
    [
        {
            'direction': [a,b,c],
            0 : 
            {
                'ord':
                [
                    {
                        'interval': [a,b],
                        'index': k,
                        'sign': True,
                        'essential': True
                    }
                ]
                ,
                'ess':,
                'rel':
            },
            1:
            
        },
    ]
    The assumption is that the list is ordered by pairs of antipodal directions.
    
    For the sign:
        True is being treated as +
        False is being treated as -
        
    '''
    interior_xpht = []
    
    for direction in range(len(exterior_xpht)):
        xpht = {}
        xph = exterior_xpht[direction]
        
        xpht['direction'] = xph['direction']
        
        for k in range(3):
            ### check ord
            ord_k = xph[k]['ord']
            new_ord_k = []
            for pair in ord_k:
                if (pair['index'] == k and pair['sign'] == True):
                    new_ord_k.append(pair)
            
            ### check ess
            ess_k = xph[k]['ess']
            new_ess_k = []
            for pair in ess_k:
                if (pair['index'] == k and pair['sign'] == True):
                    new_ess_k.append(pair)
                
            
            ### check rel
            rel_k = xph[k]['rel']
            new_rel_k = []
            for pair in rel_k:
                if (pair['index'] == 2 - k and pair['sign'] == False):
                    new_rel_k.append(pair)
                    
            xpht[k] = {
                'ord': new_ord_k,
                'ess': new_ess_k,
                'rel': new_rel_k
            }
            
        interior_xpht.append(xpht)       
    return interior_xpht
```
